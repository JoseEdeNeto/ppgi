## Importando o dataset com o pandas:

```python
iris = pd.read_csv("iris.data", names=["sepal-length", "sepal-width", "petal-length", "petal-width", "target"])
iris.head(10)
```

### Algumas informações sobre o dataset:

```python
iris.describe()
```

### sepal-length:
4.3 >= pequena < 5.5<br>
5.5 >= média < 6.7<br>
6.7 >= grande <= 7.9

### sepal-width:
2 >= pequena < 2.8<br>
2.8 >= média < 3.6<br>
3.6 >= grande <= 4.4

### petal-length:
1 >= pequena < 3<br>
3 >= média < 5<br>
5 >= grande <= 6.9

### petal-width:
0.1 >= pequena < 0.9<br>
0.9 >= média < 1.7<br>
1.7 >= grande <= 2.5

### encoding:
pequena = 1<br>
media = 2<br>
grande = 3

```python
iris.loc[(iris['sepal-length'] >= 4.3) & (iris['sepal-length'] < 5.5), 'sepal-length'] = 1
iris.loc[(iris['sepal-length'] >= 5.5) & (iris['sepal-length'] < 6.7), 'sepal-length'] = 2
iris.loc[(iris['sepal-length'] >= 6.7), 'sepal-length'] = 3

iris.loc[(iris['sepal-width'] >= 2) & (iris['sepal-width'] < 2.8), 'sepal-width'] = 1
iris.loc[(iris['sepal-width'] >= 2.8) & (iris['sepal-width'] < 3.6), 'sepal-width'] = 2
iris.loc[(iris['sepal-width'] >= 3.6), 'sepal-width'] = 3

iris.loc[(iris['petal-length'] >= 1) & (iris['petal-length'] < 3), 'petal-length'] = 1
iris.loc[(iris['petal-length'] >= 3) & (iris['petal-length'] < 5), 'petal-length'] = 2
iris.loc[(iris['petal-length'] >= 5), 'petal-length'] = 3

iris.loc[(iris['petal-width'] >= 1.7), 'petal-width'] = 3
iris.loc[(iris['petal-width'] >= 0.9) & (iris['petal-width'] < 1.7), 'petal-width'] = 2
iris.loc[(iris['petal-width'] >= 0.1) & (iris['petal-width'] < 0.9), 'petal-width'] = 1

iris

iris.describe()
```
